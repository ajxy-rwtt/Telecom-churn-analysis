```python
#making changes to the column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns)
```

```python
# Count unique values in the 'state' column
unique_states_count = df['state'].nunique()
unique_states_list = df['state'].unique()

print(f'There are {unique_states_count} unique states.')
print('Unique states:', unique_states_list)
```

```python
# Comprehensive mapping of state abbreviations to full names
full_state_mapping = {
    'KS': 'Kansas', 'OH': 'Ohio', 'NJ': 'New Jersey', 'OK': 'Oklahoma', 'AL': 'Alabama',
    'MA': 'Massachusetts', 'MO': 'Missouri', 'WV': 'West Virginia', 'RI': 'Rhode Island',
    'IA': 'Iowa', 'MT': 'Montana', 'ID': 'Idaho', 'VT': 'Vermont', 'VA': 'Virginia',
    'TX': 'Texas', 'FL': 'Florida', 'CO': 'Colorado', 'AZ': 'Arizona', 'NE': 'Nebraska',
    'WY': 'Wyoming', 'IL': 'Illinois', 'NH': 'New Hampshire', 'LA': 'Louisiana',
    'GA': 'Georgia', 'AK': 'Alaska', 'MD': 'Maryland', 'AR': 'Arkansas', 'WI': 'Wisconsin',
    'OR': 'Oregon', 'DE': 'Delaware', 'IN': 'Indiana', 'UT': 'Utah', 'CA': 'California',
    'SD': 'South Dakota', 'NC': 'North Carolina', 'WA': 'Washington', 'MN': 'Minnesota',
    'NM': 'New Mexico', 'NV': 'Nevada', 'DC': 'District of Columbia', 'NY': 'New York',
    'KY': 'Kentucky', 'ME': 'Maine', 'MS': 'Mississippi', 'MI': 'Michigan', 'SC': 'South Carolina',
    'TN': 'Tennessee', 'PA': 'Pennsylvania', 'HI': 'Hawaii', 'ND': 'North Dakota', 'CT': 'Connecticut'
}

# Apply the full mapping
df['state'] = df['state'].map(full_state_mapping).fillna(df['state'])

# Verify the change by looking at unique values
print("Unique state names now:")
print(df['state'].unique())
display(df.head())
```

```python
# Identify and drop the specific columns
# Adding errors='ignore' prevents a KeyError if the columns were already dropped
columns_to_drop = ['total_intl_minutes', 'total_intl_calls', 'total_intl_charge']
df = df.drop(columns=columns_to_drop, errors='ignore')

# Verify the columns are removed
print(f"Attempted to drop columns: {columns_to_drop}")
print("Current columns in DataFrame:")
print(df.columns.tolist())
display(df.head())
```

```python
# Reloading the data to recover dropped columns
df = pd.read_csv(f"{path}/churn-bigml-80.csv")

# Re-applying cleaning steps
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df['state'] = df['state'].map(full_state_mapping).fillna(df['state'])
df.insert(0, 'row_id', range(1, len(df) + 1))

print("Data reloaded and cleaned. All columns are now present.")
display(df.head())
```

```python
import matplotlib.pyplot as plt
df['churn'].value_counts().plot(kind='bar', figsize=(4, 3))
plt.title('Churn Distribution')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
df.groupby('international_plan')['churn'].mean().plot(kind='bar', figsize=(4, 3))
plt.title('Churn Rate by International Plan')
plt.ylabel('Churn Rate')
plt.show()
```

```python
df.groupby('customer_service_calls')['churn'].mean().plot(figsize=(4,3))
plt.title('Churn Rate vs Customer Service Calls')
plt.xlabel('Number of Calls')
plt.ylabel('Churn Rate')
plt.show()
```

```python
pivot = df.pivot_table(
    values='churn',
    index='customer_service_calls',
    columns='international_plan',
    aggfunc='mean'
)

pivot.plot(figsize=(4,3))
plt.title('Churn: Service Calls vs International Plan')
plt.ylabel('Churn Rate')
plt.show()
```

```python
output_file = 'telecom_churn_final_data.csv'
df.to_csv(output_file, index=False)
print(f"DataFrame successfully saved to '{output_file}' in your Colab environment.")
```
